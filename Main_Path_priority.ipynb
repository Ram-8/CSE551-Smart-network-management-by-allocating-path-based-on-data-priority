```python
from fabrictestbed_extensions.fablib.fablib import FablibManager as fablib_manager
fablib = fablib_manager()
slice = fablib.new_slice(name='priority_path_allocation')
```

```python
site_high_priority = 'DALL'
site_medium_priority = 'MAX'
site_low_priority = 'UCSD'
print(f'The selected sites are: High Priority - {site_high_priority}, Medium Priority - {site_medium_priority}, Low Priority - {site_low_priority}')
```

```python
client = slice.add_node(name="client", site='WASH', cores=4, ram=8, disk=20, image='default_ubuntu_20')
server = slice.add_node(name="server", site='TACC', cores=4, ram=8, disk=20, image='default_ubuntu_20')
```

```python
switch_high_priority = slice.add_node(name="switch_high_priority", site=site_high_priority)
switch_medium_priority = slice.add_node(name="switch_medium_priority", site=site_medium_priority)
switch_low_priority = slice.add_node(name="switch_low_priority", site=site_low_priority)
```

```python
client_iface_high = client.add_component(model='NIC_Basic', name='clientToSwitchHigh').get_interfaces()[0]
switch_high_iface = switch_high_priority.add_component(model='NIC_Basic', name='switchToClientHigh').get_interfaces()[0]
slice.add_l2network(name='net_high_priority', interfaces=[client_iface_high, switch_high_iface])

client_iface_med = client.add_component(model='NIC_Basic', name='clientToSwitchMed').get_interfaces()[0]
switch_med_iface = switch_medium_priority.add_component(model='NIC_Basic', name='switchMed').get_interfaces()[0]
slice.add_l2network(name='net_medium_priority', interfaces=[client_iface_med, switch_med_iface])

client_iface_low = client.add_component(model='NIC_Basic', name='clientToSwitchLow').get_interfaces()[0]
switch_low_iface = switch_low_priority.add_component(model='NIC_Basic', name='switchLow').get_interfaces()[0]
slice.add_l2network(name='net_low_priority', interfaces=[client_iface_low, switch_low_iface])
```

```python
slice.submit()
```
